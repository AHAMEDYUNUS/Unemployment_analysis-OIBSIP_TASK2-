```python
import datetime as dt
import calendar
```

```python
data2x.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
              'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']

data2x['Date'] = pd.to_datetime(data2x['Date'], dayfirst=True)
data2x['Frequency'] = data2x['Frequency'].astype('category')
data2x['Region'] = data2x['Region'].astype('category')
data2x['Month'] = data2x['Date'].dt.month
data2x['Month_int'] = data2x['Month'].apply(lambda x: int(x))
data2x['Month_name'] = data2x['Month_int'].apply(lambda x: calendar.month_abbr[x])
data2x.drop(columns= 'Month',inplace = True)
```

```python
unemployment = data2x[['States', 'Region', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']]
unemplo = unemployment.groupby(['Region', 'States'])['Estimated Unemployment Rate'].mean().reset_index()
fig = pex.sunburst(unemplo, path=['Region', 'States'], values='Estimated Unemployment Rate',
                  color_continuous_scale='Plasma', title='Unemployment rate in each region and state',
                  height=650, template='plotly_dark')
fig.show()
```

```python
data_to_plot = data2x['Estimated Unemployment Rate'].values.reshape(-1, 1)

# Create a heatmap
sns.heatmap(data_to_plot, cmap="YlGnBu", annot=True, fmt=".2f")

# Show the plot
plt.show()
```
